```python
# Leer archivo

import pandas as pd
df = pd.read_csv('fifa_eda.csv')

df.head()
```

```python
# Obtener matriz de correlacion y dejarla com heatmap

import plotly.express as px

df.describe()
```

```python
#Definicion de variables para la correlacion
df_num_vars = df[['Age','Overall','Potential','Value','Wage','International Reputation','Skill Moves','Joined','Height','Weight','Release Clause']]
df_num_vars.head()
```

```python
#Definicion de la correlacion
df_num_vars.corr()
```

```python
#Grafico de correlacion
fig = px.imshow(df_num_vars.corr(), text_auto=True)
fig.show()
```

```python
# Elemento 2: Gráfico de Correlación que responda – “Cuál es la relación entre la edad y el overall?” – explicar la salida 

fig = px.scatter(df, x = "Age", y = "Overall", title="Grafico de correlacion entre la Edad y la Valoración Total del Jugador")
fig.show()
```

Media del Overall es 66.23 y de la Edad es 25 años

Con esto se pueden inferir varias cosas.  Hasta los 25 años se tienen marcas de Overall por debajo de los 50 puntos, salvo casos aislados mas adelante. A partir de ahí valoración mínima sube y es de anotar que entre mas edad, las puntuaciones más bajas pertenecen a outlayers.  En cuanto a la valoración mas altas esta entre los 26 y 33 años, que se puede inferir que son los años más productivos en general de los jugadores.

A partir de los 34 años comienzan a disminuir el número de jugadores lo que supone que los 35 años es como la edad donde se comienzan a retirar los jugagores, y comienzan a perder valor, aunque los mínimos no bajan de 50 como pasa antes de los 25.

La reducción más fuerte de jugadores esta desde los 40 años en adelante, no habiendo mayores de 45 años.  Definitivamente la data marca 3 puntos importantes.

1. Cuando son muy jovenes falta experiencia, por tanta la valoración no es alta.
2. A partir de los 25 ya se obtiene mas experiencia y se tiene mucho vigor por lo que la valoración sube y se mantiene así hasta los 33 años.
3. A pertir de los 33 años, pesa la experiencia lo cual tiene su recomenza en valoración, pero comienza a darse las bajas de jugadores y son pocos los que juegan despues de los 35 años.

```python
#Elemento 3: Generar un gráfico de barras por club que indique el número de jugadores 

df_clubes = df.groupby('Club').size().rename('Num_Jugadores').reset_index().sort_values(by='Num_Jugadores', ascending=False)
df_clubes
```

```python
# Grafico de barras de clubs vs numero de jugadores
fig = px.bar(df_clubes, x = 'Club', y = 'Num_Jugadores', title="Gráfico de Clubes con número de Jugadores")
fig.show()
```

```python
# Elemento 4: Hacer un gráfico multipanel que indique la relación altura (height) vs skill moves, siendo la variable del panel si es zurdo o derecho 

fig = px.strip(df, x='Skill Moves', y="Height", facet_col="Preferred Foot",  title="Grafico de relacion de Skill Moves con la Altura")
fig.show()
```

```python
#Elemento 5: Un gráfico que crea importante para mostrar a los potenciales cracks

#De acuerdo a las estadísticas que ya hemos visto, y basandonos en puntuaciones de variables mínimas que sean la media + la desviación standard, deberíamos ver potenciales cracks.

#Potential > 78
# Edad 26-33 años
# Overall >90
# International Reputation >1.6
# Release Clause > 70000

#Con estos datos nos refina la data y nos devuelve los mejores 23 jugadores de la base de datos.

df_cracks = df[(df['Potential'] > 78 ) & (df['Age'] > 25) & (df['Age'] <= 33) & (df['Overall'] > 88) & (df['International Reputation'] > 1.6) & (df['Release Clause'] > 70000)].sort_values(by='Age', ascending=True)
df_cracks
```

```python
#Gráfico de los cracks
fig = px.strip(df_cracks, x="Name", y="Overall", color="Age",title="Grafico de los Cracks de la liga")
fig.show()
```
