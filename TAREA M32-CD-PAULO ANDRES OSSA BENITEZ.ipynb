```python
# Leer archivo

import pandas as pd
df = pd.read_csv('fifa_eda.csv')

df.head()
```

```python
# Obtener matriz de correlacion y dejarla com heatmap

import plotly.express as px

df.describe()
```

```python
#Definicion de variables para la correlacion
df_num_vars = df[['Age','Overall','Potential','Value','Wage','International Reputation','Skill Moves','Joined','Height','Weight','Release Clause']]
df_num_vars.head()
```

```python
#Definicion de la correlacion
df_num_vars.corr()
```

```python
#Grafico de correlacion
fig = px.imshow(df_num_vars.corr(), text_auto=True)
fig.show()
```

```python
# Elemento 2: Grfico de Correlacin que responda  Cul es la relacin entre la edad y el overall?  explicar la salida 

fig = px.scatter(df, x = "Age", y = "Overall", title="Grafico de correlacion entre la Edad y la Valoracin Total del Jugador")
fig.show()
```

Media del Overall es 66.23 y de la Edad es 25 aos

Con esto se pueden inferir varias cosas.  Hasta los 25 aos se tienen marcas de Overall por debajo de los 50 puntos, salvo casos aislados mas adelante. A partir de ah valoracin mnima sube y es de anotar que entre mas edad, las puntuaciones ms bajas pertenecen a outlayers.  En cuanto a la valoracin mas altas esta entre los 26 y 33 aos, que se puede inferir que son los aos ms productivos en general de los jugadores.

A partir de los 34 aos comienzan a disminuir el nmero de jugadores lo que supone que los 35 aos es como la edad donde se comienzan a retirar los jugagores, y comienzan a perder valor, aunque los mnimos no bajan de 50 como pasa antes de los 25.

La reduccin ms fuerte de jugadores esta desde los 40 aos en adelante, no habiendo mayores de 45 aos.  Definitivamente la data marca 3 puntos importantes.

1. Cuando son muy jovenes falta experiencia, por tanta la valoracin no es alta.
2. A partir de los 25 ya se obtiene mas experiencia y se tiene mucho vigor por lo que la valoracin sube y se mantiene as hasta los 33 aos.
3. A pertir de los 33 aos, pesa la experiencia lo cual tiene su recomenza en valoracin, pero comienza a darse las bajas de jugadores y son pocos los que juegan despues de los 35 aos.

```python
#Elemento 3: Generar un grfico de barras por club que indique el nmero de jugadores 

df_clubes = df.groupby('Club').size().rename('Num_Jugadores').reset_index().sort_values(by='Num_Jugadores', ascending=False)
df_clubes
```

```python
# Grafico de barras de clubs vs numero de jugadores
fig = px.bar(df_clubes, x = 'Club', y = 'Num_Jugadores', title="Grfico de Clubes con nmero de Jugadores")
fig.show()
```

```python
# Elemento 4: Hacer un grfico multipanel que indique la relacin altura (height) vs skill moves, siendo la variable del panel si es zurdo o derecho 

fig = px.strip(df, x='Skill Moves', y="Height", facet_col="Preferred Foot",  title="Grafico de relacion de Skill Moves con la Altura")
fig.show()
```

```python
#Elemento 5: Un grfico que crea importante para mostrar a los potenciales cracks

#De acuerdo a las estadsticas que ya hemos visto, y basandonos en puntuaciones de variables mnimas que sean la media + la desviacin standard, deberamos ver potenciales cracks.

#Potential > 78
# Edad 26-33 aos
# Overall >90
# International Reputation >1.6
# Release Clause > 70000

#Con estos datos nos refina la data y nos devuelve los mejores 23 jugadores de la base de datos.

df_cracks = df[(df['Potential'] > 78 ) & (df['Age'] > 25) & (df['Age'] <= 33) & (df['Overall'] > 88) & (df['International Reputation'] > 1.6) & (df['Release Clause'] > 70000)].sort_values(by='Age', ascending=True)
df_cracks
```

```python
#Grfico de los cracks
fig = px.strip(df_cracks, x="Name", y="Overall", color="Age",title="Grafico de los Cracks de la liga")
fig.show()
```
